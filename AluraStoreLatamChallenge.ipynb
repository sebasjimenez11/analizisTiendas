### Importaci√≥n de datos



```python
import pandas as pd

# URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Lectura
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# A√±adir columna para identificar de qu√© tienda viene cada fila
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los datos
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

df
```

#1. An√°lisis de facturaci√≥n



```python
df['Facturacion'] = df['Precio'] + df['Costo de env√≠o']
facturacion_total = df.groupby('Tienda')['Facturacion'].sum().sort_values()
print("Facturaci√≥n total por tienda:\n", facturacion_total)
```

# 2. Ventas por categor√≠a

```python
ventas_categoria = df.groupby(['Tienda', 'Categor√≠a del Producto'])['Producto'].count()
print("Ventas por categor√≠a:\n", ventas_categoria)
```

# 3. Calificaci√≥n promedio de la tienda


```python
clasificacion_prom = df.groupby('Tienda')['Calificaci√≥n'].mean().sort_values()
print("Calificaci√≥n promedio por tienda:\n", clasificacion_prom)
```

# 4. Productos m√°s y menos vendidos

```python
mas_vendidos = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad')
mas_vendidos_sorted = mas_vendidos.sort_values(['Tienda', 'Cantidad'], ascending=[True, False])
print("Productos m√°s y menos vendidos por tienda:")
for tienda in df['Tienda'].unique():
    temp = mas_vendidos_sorted[mas_vendidos_sorted['Tienda'] == tienda]
    print(f"\n{tienda}")
    print("M√°s vendido:", temp.iloc[0]['Producto'], "-", temp.iloc[0]['Cantidad'])
    print("Menos vendido:", temp.iloc[-1]['Producto'], "-", temp.iloc[-1]['Cantidad'])
```

# 5. Env√≠o promedio por tienda

```python
envio_promedio = df.groupby('Tienda')['Costo de env√≠o'].mean().sort_values()
print("üì¶ Env√≠o promedio por tienda:\n", envio_promedio)
```

# **An√°lisis final para decidir qu√© tienda vender**

```python
resumen = pd.DataFrame({
    'Facturaci√≥n Total': facturacion_total,
    'Calificaci√≥n Prom.': clasificacion_prom,
    'Costo Env√≠o Prom.': envio_promedio,
    'Total Ventas': df.groupby('Tienda').size()
}).sort_values('Facturaci√≥n Total')

print("Resumen de m√©tricas clave:\n", resumen)
```

# **Recomendaci√≥n final**

```python
tienda_menos_rentable = resumen.index[0]
print(f"\nRecomendaci√≥n: Vender o reestructurar {tienda_menos_rentable}, ya que presenta la facturaci√≥n m√°s baja, baja calificaci√≥n y bajo volumen de ventas.")
```
